```python
index_list = []
for i in df['tweet']:
    try:
        isReliable, textBytesFound, details = cld2.detect(str(i))
        if(isReliable == True and details[0][0] == 'ENGLISH'):
            index_list.append(1)
        else:
            index_list.append(0)
    except:
        index_list.append(-500);

print(index_list)
```

```python
# let's convert this list into filtered dataframe
filtered_list = [number for number in range(0,len(index_list)) if index_list[number] == 0]
print("Index of rows which contain 0 or -500:" , filtered_list)
print("length of filterd list:", len(filtered_list))
Filter_df  = df[df.index.isin(filtered_list)]
```

```python
Filter_df.reset_index(inplace=True, drop=True)
Filter_df
```

```python
df.drop(labels=None, axis=0, index=filtered_list, columns=None, level=None, inplace=True, errors='raise')
```
